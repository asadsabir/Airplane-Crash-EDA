```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('./Airplane_Crashes_and_Fatalities_Since_1908_t0_2023.csv',encoding='latin-1',parse_dates=True)



df.drop(['cn/ln','Registration','Flight #','Time','Summary','Location','Operator','AC Type','Route'],axis=1,inplace=True)
df.head() 
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
sns.pairplot(df)
```

```python
sns.histplot(df['Ground']);
```

```python
len(df[df['Ground'] != 0])/len(df)
```

```python
sns.histplot(df[df['Ground'] != 0]['Ground'])
```

```python
sns.scatterplot(df[df['Ground'] != 0],x='Date',y='Ground')
```

```python
len(df[df['Aboard'] < (df['Aboard Crew'] + df['Aboard Passangers'])])
```

```python
df[df['Aboard'] != (df['Aboard Crew'] + df['Aboard Passangers'])]
```

```python
df[df['Fatalities'] != (df['Fatalities Crew'] + df['Fatalities Passangers'] + df['Ground'])]
```

```python
df['Fatalities'] = df['Fatalities Crew'] + df['Fatalities Passangers'] + df['Ground']
df['Aboard'] = df['Aboard Crew'] + df['Aboard Passangers']
```

```python

df['Year'] = df['Date'].dt.year

by_year = df.drop('Date',axis=1).groupby('Year').agg('mean')
```

```python
count = df.drop('Date',axis=1).groupby('Year').agg('count')
```

```python
by_year.head()
```

```python
by_year['Incidents'] = count['Aboard']
```

```python
by_year.head()
```

```python
len(by_year[by_year['Fatalities'] == 0])
```

```python
sns.scatterplot(by_year,x='Year',y='Incidents')
plt.grid(True)
```

```python
by_year['Aboard Fatalities %'] = (by_year['Fatalities Crew'] + by_year['Fatalities Passangers']) * 100 / by_year['Aboard']
```

```python
sns.scatterplot(by_year,x='Year',y='Aboard Fatalities %')
```

```python
sns.scatterplot(by_year,x='Year',y='Incidents',hue='Aboard Fatalities %')
plt.grid(True)
```

```python
for col in by_year.columns:
    if col != 'Year':
        sns.scatterplot(by_year,x='Year',y=col)
        plt.show()
```

```python
by_year.head()
```

```python
by_year['Number of Fatalities on the ground'] = (by_year['Ground']*by_year['Incidents'])

sns.scatterplot(by_year[by_year['Ground'] < 80],x='Year',y = 'Number of Fatalities on the ground')
```
